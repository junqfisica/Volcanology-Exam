```python
%matplotlib inline
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
# initial parameter
xmax = 10.0
nx = 200
dx = xmax/(nx-1)
x = np.linspace(0,xmax,nx)

l = 20*dx
k = 2*np.pi/l
f = np.sin(k*x)
```

```python
plt.plot(x,f)
plt.xlim(0,xmax)
plt.grid()
```

```python
norder = np.zeros(nx)
ader = np.zeros(nx)

for i in range(1, nx-1):
    norder[i] = (f[i+1] - f[i-1])/(2*dx)

ader = k * np.cos(k*x)
ader[0] = 0.
ader[nx-1] = 0.

rms = np.sqrt(np.mean(norder-ader)**2)
```

```python
plt.plot(x,norder,  '+', label='Numerical Derivative')
plt.plot(x,ader, label='Analitical Derivative')
plt.plot(x,norder-ader, label='Derivative Difference')
plt.xlim(0,xmax)
plt.grid()
plt.legend()
```
